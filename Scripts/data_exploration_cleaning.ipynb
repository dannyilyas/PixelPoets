```python
def search_keyword(df, word):
  keyword_rows = df[df['s2_caption'].str.contains(word, case=False, na=False)]
  return keyword_rows
```

```python
import string
# Remove 'figure'
def remove_figure(text):
    pattern = r'Figure \d+'
    return re.sub(pattern, '', text)

def remove_punctuation(input_string):
    # Make a regular expression that matches all punctuation
    regex = re.compile('[%s]' % re.escape(string.punctuation))
    # Use the regex
    return regex.sub('', input_string)

def remove_figure2(text):
    pattern = r'Fig \d+'
    return re.sub(pattern, '', text)

def remove_figure3(text):
    pattern = r'Figures \d+'
    return re.sub(pattern, '', text)
def remove_capital(text):
    pattern = r'\b[A-Z]\b'
    return re.sub(pattern, '', text)

df['s2_caption_cleaned'] = df['s2_caption'].apply(remove_figure)
df['s2_caption_cleaned'] = df['s2_caption_cleaned'].apply(remove_punctuation)
df['s2_caption_cleaned'] = df['s2_caption_cleaned'].apply(remove_figure2)
df['s2_caption_cleaned'] = df['s2_caption_cleaned'].apply(remove_figure3)
df['s2_caption_cleaned'] = df['s2_caption_cleaned'].apply(remove_capital)
df['s2_caption_cleaned']
```

```python
# Save as new csv file
path = '/content/drive/My Drive/CSC490/CSC490 Project/cleaned.csv'
df.to_csv(path, index=False)
```
